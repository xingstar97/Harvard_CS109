```python
# Activation Function
def sigmoid(x):
    return 1/(1 + np.exp(-x))

# Neural Network model 
def nn_one_layer(input_dim, output_dim, num_hidden_nodes, x_train, y_train, WL0=None, WL1=None): 
    # Intial weights
    if WL0 is None: 
        WL0 = np.ones((num_hidden_nodes, input_dim + 1))
    
    if WL1 is None: 
        WL1 = np.ones((output_dim, num_hidden_nodes + 1))
    
    # input layper
    ones = np.ones((len(x_train), 1))
    L0 = x_train
    L0 = np.append(L0, ones, axis=1)
    
    # hidden layer
    a0 = np.dot(WL0, L0.T)
    z0 = {'sigmoid': sigmoid(a0)}
    
    # output layer
    ones = np.ones((len(y_train), 1))
    z0['sigmoid'] = np.append(z0['sigmoid'].T, ones, axis=1).T
    
    # affine
    a1 = {'sigmoid': np.dot(WL1, z0['sigmoid'])}
    z1 = a1
    
    return z1
```

```python
# Plot the model
def plot_nn_models(result_df, title):
    result_df = result_df.sort_values(by='x_train')
    
    fig, ax = plt.subplots(1, 1, figsize=(11, 7))
    
    ax.scatter(result_df.x_train, result_df.y_train, color='k', marker='*', lw=1, label='Actual Values')
    ax.plot(result_df.x_train, result_df.y_pred_train_sigmoid, '--', lw=4, label='Predicted Values')
    ax.axhline(y=0.5, linestyle='-.')
    
    ax.set_title(title, fontsize=16)
    ax.set_xlabel('$x$', fontsize=16)
    ax.set_ylabel('$y$', fontsize=16)
    ax.tick_params(labelsize=16)
    ax.legend(fontsize=16, loc='best')
    
    return
```

```python
# train set
x_train, y_train = df_1step['x'].values.reshape(-1,1), df_1step['y'].values

# Initial Weights
input_dim = 1
output_dim = 1
num_hidden_nodes = 1

# Calibrated weights
WL0 = np.array([[20, -40]])
WL1 = np.array([[1, 0]])

# Neural Network Model 
nn_1step = nn_one_layer(input_dim, output_dim, num_hidden_nodes, x_train, y_train, WL0, WL1)

result_1step = pd.DataFrame({'x_train': x_train[:, 0],
                          'y_train': y_train,
                          'y_pred_train_sigmoid': nn_1step['sigmoid'][0]
                         })
plot_nn_models(result_1step, 'One Step Function with Calibrated Weights')
```

```python
# Loss function 
def loss(result_df): 
    return sum((result_df.y_pred_train_sigmoid - result_df.y_train)**2)

loss_1step = loss(result_1step)
loss_1hump = loss(result_1hump)
loss_2hump = loss(result_2hump)

print("Loss functoin for one step fuction: %f" % loss_1step)
print("Loss functoin for one hump fuction: %f" % loss_1hump)
print("Loss functoin for two equal humps fuction: %f" % loss_2hump)
```

```python
# inline function for f(x)
f = lambda x: np.sin(x) + np.cos(x)

# inline function for df/dx
der_f_x = lambda x: np.cos(x) - np.sin(x)

x_n_minus_1 = 1
f_x_min = -1
x_vals = [x_n_minus_1]
errors = []

l = 0.01 
error = 1
step = 0
threshold = 0.0001

while True:
    gradient_eval = der_f_x(x_n_minus_1) 
    x_n = x_n_minus_1 + l*gradient_eval 
    f_eval = f(x_n) 
    error = f_eval - f_x_min 
    
    errors.append(error) 
    x_vals.append(x_n) 
    
    x_n_minus_1 = x_n 
    step += 1
    
    if abs(f_eval - f(x_vals[step-1])) < threshold: break
    
print('Number of steps = ', step)

figure, ax = plt.subplots(1, 2, figsize=(18,6))
ax[0].plot(range(len(x_vals)), x_vals)
ax[0].set_xlabel('step')
ax[0].set_ylabel('x_min')
ax[1].set_title('Gradient Descent')

ax[1].plot(range(len(errors)), errors)
ax[1].set_xlabel('step')
ax[1].set_ylabel('error')
ax[1].set_title('Gradient Descent')
```

```python
# your code here
pima_df = pd.read_csv("data/pima-indians-diabetes.csv")
print(pima_df.describe())
print(pima_df.dtypes)
columns = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

##remove special character of the outcome 
pima_df['Outcome'] = pima_df.Outcome.str.extract('(\d+)')
pima_df = pima_df[(pd.to_numeric(pima_df.Outcome) == 0) | (pd.to_numeric(pima_df.Outcome) == 1)]

##check if any NA values
print(pima_df.isnull().any().any())
```

```python
pima_df_clean = pima_df.copy()

pima_df_clean = pima_df_clean[(pima_df_clean.Glucose!=0) & (pima_df_clean.BloodPressure!=0) & \
        (pima_df_clean.SkinThickness !=0) & (pima_df_clean.Insulin!=0) & \
        (pima_df_clean.BMI !=0) & (pima_df_clean.DiabetesPedigreeFunction !=0) ]


print(pima_df_clean.head())
print(pima_df_clean.shape)
```

In the outcome column, there are values with special character '\' etc so we use the regex to extract the numberic value and remove the values not in [0,1].

There are no NA values in the dataset but there are 0 values in Glucose, BloodPressure, SkinThickness, Insulin, BMI and DiabetesPedigreeFunction which does not make sense. Therefore, those are disguised missing data.

**2.2** 

```python
# your code here
def evaluate_model(x,y,test_size, random_state):
    X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=test_size, random_state=random_state)
    clf = LogisticRegressionCV().fit(X_train,y_train)
    return clf
```

```python
X_train, X_test, y_train, y_test = train_test_split(pima_df_clean.drop(columns = 'Outcome'),\
                           pima_df_clean[['Outcome']], test_size=0.25, random_state=9001)
```

```python
clf_clean = evaluate_model(pima_df_clean.drop(columns = 'Outcome'),\
                           pima_df_clean[['Outcome']], test_size=0.25, random_state=9001)

print("Accuracy for training data set:",clf_clean.score(X_train,y_train))
print("Accuracy for testing data set:",clf_clean.score(X_test,y_test))
```

**2.3** 

```python
# your code here
#fill mean to the 0 values 
pima_df_mean = pima_df.copy()
imp_mean = Imputer(missing_values=0, strategy='mean')
pima_df_mean[columns] = imp_mean.fit_transform(pima_df_mean[columns])
print(pima_df_mean.shape)

#evaluate the logistic regression results
clf_mean = evaluate_model(pima_df_mean.drop(columns = 'Outcome'),\
                     pima_df_mean[['Outcome']],test_size=0.25, random_state=9001)
print("Accuracy for training data set:",clf_mean.score(X_train,y_train))
print("Accuracy for testing data set:",clf_mean.score(X_test,y_test))
```

**2.4** 

```python
# your code here
# We are going to use linear regression to fill the missing vaule

from sklearn.linear_model import LinearRegression

pima_df_model = pima_df.copy()

for col in columns:
    valid_x = pima_df_clean.drop(columns=[col,"Outcome"])
    valid_y = pima_df_clean[col]
    reg = LinearRegression().fit(valid_x, valid_y)
    fill_value = reg.predict(pima_df_model[pd.to_numeric(pima_df_model[col])==0].drop(columns=[col,"Outcome"]))
    fill_value = fill_value + np.random.normal(0,np.sqrt(mean_squared_error(valid_y, reg.predict(valid_x))), size =len(fill_value))    
    pima_df_model.loc[pd.to_numeric(pima_df_model[col])==0,col] = fill_value
    
    
```

```python
print(pima_df_model.shape)
clf_model = evaluate_model(pima_df_model.drop(columns = 'Outcome'),\
                           pima_df_model[['Outcome']],test_size=0.25, random_state=9001)
print("Accuracy for training data set:",clf_model.score(X_train,y_train))
print("Accuracy for testing data set:",clf_model.score(X_test,y_test))
```
